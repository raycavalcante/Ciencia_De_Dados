```python
import pandas as pd
```

```python
produtos_df = pd.read_csv('arquivos_pandas/Contoso - Cadastro Produtos.csv', sep=';', encoding='latin-1')
cliente_df = pd.read_csv('arquivos_pandas/Contoso - Clientes.csv', sep=';', encoding='latin-1')
lojas_df = pd.read_csv('arquivos_pandas/Contoso - Lojas.csv', sep=';', encoding='latin-1')
promocoes_df = pd.read_csv('arquivos_pandas/Contoso - Promocoes.csv', sep=';', encoding='latin-1')
vendas_df = pd.read_csv('arquivos_pandas/Contoso - Vendas - 2017.csv', sep=';', encoding='latin-1')
display(vendas_df)
```

```python
# exibindo apenas colunas espeficica 
lista_colunas = ['ID Produto','Quantidade Vendida','Quantidade Devolvida']
produtos_quantidade = vendas_df[lista_colunas]
display(produtos_quantidade)
```

```python
display(cliente_df)
```

```python
# dataframe somente com colunas especificas
# Dizemas o dataframe irar receber o mesmo dataframe  porem com colunas especificas a serem mostradas
cliente_df = cliente_df[['ID Cliente','E-mail']]
produtos_df = produtos_df[['ID Produto','Nome do Produto']]
lojas_df = lojas_df[['ID Loja','Nome da Loja']]
display(lojas_df)
```

```python
# juntar (merge) as tabelas que se relacionam
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(cliente_df, on='ID Cliente')
display(vendas_df)
```

```python
display(vendas_df)
```

```python
# qual cliente comprou mais vezes
frequencia_clientes = vendas_df['E-mail'].value_counts()
display(frequencia_clientes)
```

```python
# gerando grafiico
frequencia_clientes.plot()
```
