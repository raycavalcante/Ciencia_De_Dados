```python
import numpy as np
```

```python
# gerando numeros aleatorios l
rng = np.random.default_rng()
print('Numeros aleatorios: ')
print(rng.random(4) * 10)
```

```python
# exercicio 
# gere 500.000 numeros aleatorios 
num_aleatorio = np.random.default_rng()
print(num_aleatorio.random(2) * 500)
# mostre o maior e menor do array 
array_brincando = np.array(rng.random(500_000) * 4_000)
menor = np.min(array_brincando)
print('Menor valor: ',menor)
maior = np.max(array_brincando)
print('Maior valor: ',maior)
# mostre a media dos soma_valores
media = np.mean(array_brincando)
print('Média: ',media)
```

```python
# EXER 
# crie um array de 1 diemensão,. preencha com 900 mil registro
# variando de 20 a 15000 
valores_array = np.array(np.random.randint(1, 1000, 900000))
print(valores_array) 
```

```python
# mostre a média de desses volores
media_array = np.sum(valores_array)
print(media_array)
```

```python
# mostre o primeiro e ultimo valor
print('Primeiro valor:')
print(valores_array[0])
print('Ultimo valor: ')
print(valores_array[-1])
```

```python
# mostre o maior e menor valor
maior_valor = np.max(valores_array)
print('Maior valor: ',maior_valor)
menor_valor = np.min(valores_array)
print('Maior valor: ',menor_valor)
```

```python
# mostre o valor em ordem decrescente
print('Array inicial: ')
print(valores_array)

print('Array em ordem dexrescente: ')
print(np.sort(valores_array)[::-1])
```

```python
# conta os valores unicos e suas repetições 
array_exer = np.array(np.random.randint(20, 15_000, 900_000))

valores_unicos, contagem = np.unique(array_exer, return_counts=True)
for valor , cont in zip(valores_unicos, contagem):
    print(f'Valor {valor} - contagem {cont}')
```
